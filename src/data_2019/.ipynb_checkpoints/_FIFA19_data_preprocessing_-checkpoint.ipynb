#### Drop  *Unnamed: 0*  and  *Special*  attributes (useless for our application)
*Unnamed: 0*     $\leftarrow$  player ID used by FIFA18 system;<br>

#### Check for null values: 

```python
data.isnull().sum()
```

We can observe a lot of attributes with **missing** (*null*) values. <br>
$\rightarrow$  For the position attributes, obviosly, not all the players can play in all the position (*think about the Goalkeeper*); <br>
$\rightarrow$  Also, not all the players have a club (*Club* attribute); \[solved below\]

#### Fill the *Club* missing values with the *Free Agent* label:

```python
# data.loc[data['Club'].isnull()]
data['Club'].fillna('Free Agent', inplace = True)
data['Club Logo'].fillna('Free Agent', inplace = True)
```

```python
data.loc[data['Club'].isnull()]
#data.loc[data['Club Logo'].isnull()]
```

*Club* NaN values problem solved, now we have *Free Agent* players.

```python
NaN_attr = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 
             'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
             'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for attr in NaN_attr:
    data[attr].fillna(1, inplace = True)
```

```python
data.isnull().sum()
```

#### We can still observe 48 players with NaN values on a great subset of attributes. We are going to remove them fom the dataset.

```python
data.dropna(subset = ['Stamina', 'GKReflexes', 'Position', 'Jersey Number'],  inplace = True)
```

```python
data.isnull().sum()
```

#### We have still NaN values for:
- *Loaned From* attribute. We will fill the filed with *Not loaned*;
- *Joined* attribute. We will drop this players.
- *Contract Valid Until* attribute. We will fill the fields with *Undefined*.

```python
data['Loaned From'].fillna('NOT LOANED', inplace = True)
```

```python
data.isnull().sum()
```

```python
data.to_csv()
```

 #### Download the images for offline use, using the links in the  dataset

```python
from re import compile
from urllib.request import urlretrieve

players_photo    = data_base_path + "PlayersPictures/"
nations_pictures = data_base_path + "NationsPictures/"
clubs_pictures   = data_base_path + "ClubsPictures/"
```

Photo working link: 
        https://cdn.sofifa.org/players/4/19/158023.png

Nationality flag link:
        https://cdn.sofifa.org/flags/52.png
        
Club Logo link:
        https://cdn.sofifa.org/teams/2/light/241.png

```python
nations_set = set()
clubs_set   = set()
for row in data[['ID', 'Photo', 'Nationality', 'Flag', 'Club', 'Club Logo']].itertuples():
    ID           = row[1]
    player_photo = row[2]
    nationality  = row[3]
    nation_flag  = row[4]
    club         = row[5]
    club_logo    = row[6]
    
    club_logo_id   = club_logo.split('/')[-1].replace('.png', '')
    nation_flag_id = nation_flag.split('/')[-1].replace('.png', '')
    # print(f"ID: {ID} \t club_logo: {club_logo}; \t club_logo_id: {club_logo_id}")
    
    if (nation_flag not in nations_set):
        urlretrieve(nation_flag, nations_pictures + nation_flag_id + '.png')
    
        nations_set.add(nations_set)
        
    break
    
    if (club not in clubs_set):
        clubs_set.add(club)
        
        # get the file name from the club logo link
        name_pattern = '[0-9]+.png'
        club_name    = compile(name_pattern).findall(club_logo)
        
        urlretrieve(club_logo, clubs_pictures + club_name[0])
    
    
    # get the file name from the photo link
    name_pattern = '[0-9]+.png'
    player_name  = compile(name_pattern).findall(player_photo)
     
    urlretrieve(player_photo, players_photo + player_name[0]) 
```
